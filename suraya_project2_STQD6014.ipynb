## 4.4 Objective 1: To examine the correlation between the annual expenditure per customer and the duration of customer's membership.

```python
## -- To explore the relationship between Length of Membership vs Yearly Amount Spent -- ##
sns.lmplot(x='Length of Membership',y='Yearly Amount Spent',data=mydata,
           scatter_kws={'color':'magenta'}, line_kws={'color':'blue'})
```

The visualization displayed above reveals a noticeable trend where the length of customer membership directly influences the amount spent annually. The prominent pattern suggest that customers who have been members for a longer duration tend to spend more each year compared to those with shorter membership durations.

We can now further analyze the connection between the duration of customers' membership and the annual expenditure using linear regression. The hypothesis for objective 1 would be:


> Null hypothesis: There is no relationship between length of customers' membership and yearly annual spend per customer.


> Alternative hypothesis: There is a relationship between length of customers' membership and yearly annual spend per customer.

Additionally, the linear regression model equation formulated for objective 1 is outlined as follow:


> Yearly amount spend on customer ~ Length of membership










```python
## -- Building Linear Regression Model -- ##

import statsmodels.api as sm

X = mydata['Length of Membership']
y = mydata['Yearly Amount Spent']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

The analysis suggests that the p-value associated with the relationship between the length of customers' membership and their yearly expenditure per customer is less than the conventional significance level of 0.05. This finding implies that there is sufficient evidence to reject the null hypothesis. Therefore, we affirm the existence of a correlation between the length of customers' membership and their yearly expenditure per customer.

Additionally, upon reviewing the coefficients provided in the OLS Regression Results above, it is evident that the intercept coefficient stands at 272.37. This suggests that the annual expenditure for a customer without membership is RM 272.37.

Furthermore, we observed that the coefficient associated with the length of membership is 64.28. This indicates that for every additional unit increase in the length of membership, there is a corresponding rise of RM 64.28 in the yearly amount spent per customer. For instance, if a customer were to extend their membership by two more years, it is predicted that they would spend an additional amount of RM 128.56.

Based on this coefficient, the prediction equation can be formulated as follows:


> Yearly amount spent (RM) = 64.28(Length of customers' membership + 272.37

Having analyzed the scatter plots and assessed the linearity relationship between the yearly amount spent and both the time spent on the app and website, the next step is to **partition the dataset into training and testing sets**. This division is essential for building and evaluating predictive models. In this process, the dataset is divided into two subsets: the training set, which is used to train the model, and the testing set, which is used to evaluate the model's performance. By splitting the data, we can assess how well the model generalizes to unseen data and avoid overfitting. The variables will be set as follows:

> y = Yearly Amount Spent


> X = Time spent on app, Time spent on web

We can now further analyze the connection between the annual expenditure and time spend both on app and website using multilinear regression. The hypothesis for objective 2 would be:


> Null hypothesis: There is no relationship between yearly amount spend and the time spend on either web or app.


> Alternative hypothesis: There is a relationship between yearly amount spend and the time spend on either web or app.

Additionally, the multilinear regression model equation formulated for objective 2 is outlined as follow:


> Yearly amount spend on customer ~ Time spend on app + Time spend on website





```python
## -- To get column names in dataset -- ##
x = mydata.keys()
print(x)
```

```python
## -- Setting up the variables -- ##
y = mydata['Yearly Amount Spent']

X = mydata[['Time on App', 'Time on Website']]
```

```python
## -- Splitting the data into training and testing sets -- ##
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3 ,random_state=101)
```

### 4.4.1 Training the model

```python
## -- Training the model -- ##
import statsmodels.api as sm

X_train = sm.add_constant(X_train)

# Fit the linear regression model
model1 = sm.OLS(y_train, X_train).fit()

# Print the summary
print(model1.summary())
```

### 4.4.1 Predicting Test Data

```python
## -- Predicting Test Data -- ##
import statsmodels.api as sm

# Add a constant term to the training features
X_train_with_const = sm.add_constant(X_train)

# Fit the linear regression model using statsmodels
model2 = sm.OLS(y_train, X_train_with_const).fit()

# Add a constant term to the test features
X_test_with_const = sm.add_constant(X_test)

# Make predictions using lm.predict
predictions = model2.predict(X_test_with_const)

# Print the summary
print(model2.summary())
```

```python
sns.distplot((y_test-predictions),bins=50)
```

Based on the regression summary for both trained and predicted data suggest that the multilinear regression model developed has performed well in capturing the underlying relationships in the data. The term "good fit" implies that the model's predictions closely match the actual values of the target variable.

The mention of the histogram of residuals suggests that the residuals, which are the differences between the actual and predicted values, exhibit a nearly smooth bell-shaped distribution. A bell-shaped histogram is characteristic of a normal distribution, indicating that the residuals are approximately normally distributed. This is an important assumption of linear regression, as it implies that the errors are random and not biased towards any particular direction.

Overall, the description suggests that the linear regression model is reliable and provides accurate predictions, as evidenced by the normally distributed residuals.

### 4.4.3 Analyzing the regression summary

```python
## -- Getting the p-value of the model -- ##
print(model1.pvalues)
```

Upon reviewing the summary, it is notable that the p-value associated with time spend on the app is below 0.05, leading us to conclude that there exisist a significant relationship between the yearly amount spend per customer and the time they spent on the app.

Conversely, the p-value for time spent on the web exceeds 0.05, indicating the absence of a discernible relationship between the yearly amount spend per customer and the time they spent on the web.

```python
## -- Getting the coefficients of the model -- ##

print(model1.params)
```

Upon examining the coefficients presented earlier, we can infer that the intercept coefficient is 111.81. This value represents the anticipated yearly expenditure of customers who do not engage with either the app or website. Essentially, if a customer does not spend any time interacting with the online platforms, their projected annual spending is estimated to be RM 111.81.In other words, the anticipated yearly spending of RM 111.81 is attributed to purchases made through physical stores in the absence of any online activity.

Additionally, we observe that the coefficient associated with the time spent on the app is 39.05. This implies that for each unit increase in the amount of time a customer spends on the app, we anticipate a corresponding increase of RM 39.05 in their yearly expenditure. In simpler terms, as customers allocate more time to using the app, we expect their annual spending to rise by an average of RM 39.05.

Finally, the coefficient attributed to the time spent on the website is -2.24. This suggests that for every additional unit of time spent on the website by a customer, we anticipate a decrease of RM 2.24 in their yearly expenditure. In essence, as customers allocate more time to browsing the website, we expect their annual spending to decrease by an average of RM 2.24.

Given the explanation of the coefficients provided above, we can construct the prediction model as follows:
> Yearly amount spend = 39.05(Time spend on app) - 2.24(Time spend on website) + 111.81.
